This was challenging to say the least. In this reinforment learning example, I was aiming to balance an inverted pendulum. The state values describe the cart's position, it's velocity, objective pole angle, & the velocity trailing the pole's tip. This agent can take actions like moving the cart left or right to influence the pole's position. The objective is to create an agent that learns through trial and error to solve the problem effectively.


Usage of Deep Q-Learning or DQN algorithm was used here, which is to combine Q-learning with neural networks. By using the feature known as expereince replay in which the agent stores it's past exepreince (states, actions, rewards, & next state) in it's memory, then later samples it randomely for training. This method is to help reduce the correlation between consecutive actions, making it's learning more stable & efficient.

I set my discount factor to 0.95 as an attempt to determine how much the agent values futures rewards compared to immediate ones. A reading of the higher values encourage long-term planning, while a lower value focuses on short-term gains.

In order to handle large state spaces, the DQN algorithm replaces the traditional Q-table with a neural network. The network then approximates it's Q-values with use of the Bellman equation, making the process more scalable & efficient. Once the neural network takes the current state as input & then predicts Q-values for each possible action, thus allowing the agent to choose the best action.

The learning rate used was another critical factor. With the default learning rate of 0.001, my algorithm managed to converge at 125 runs. However, when the learning rate was reduced to 0.0001, it took 192 runs, and when increased to 0.01, the simulation failed to converge even after 784 runs.It's possible it may have converged at a later run, but at that point it had been running for nearly 45 mins, as it gets slower in it's processing. A learning rate that is too high can cause the algorithm to oscillate without converging, while a lower rate slows down learning. I then introduced a decay factor to gradually reduce the learning rate over an extended period of time to help balance faster initial learning with fine tuning later.

Due to the complexity of the problem and the need for extensive testing, many results remained inconclusive. System errors and the time required for long training sessions further complicate the evaluation. Nonetheless, this example highlights the potential of reinforcement learning and the DQN algorithm in solving control tasks like the CartPole problem.
